```python
# Step 6
df_merged['average_price'] = df_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_merged.loc[df_merged['average_price'] >= 10 , 'spending_flag'] = 'High spender'
```

```python
df_merged.loc[df_merged['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_merged['spending_flag'].value_counts()
```

```python
# Step 7
df_merged['order_frequency'] = df_merged.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
df_merged.head(60)
```

```python
df_merged.loc[df_merged['order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
df_merged.loc[(df_merged['order_frequency'] > 10) & (df_merged['order_frequency'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
df_merged.loc[df_merged['order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
df_merged['order_frequency_flag'].value_counts()
```
